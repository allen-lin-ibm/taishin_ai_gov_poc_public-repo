### Client 擇一綁定 Project 或 Space
**（兩者不能同時設定）**

#### Link default project

```python
client.set.default_project(PROJECT_ID)

print(f"✅ Default project set to: {PROJECT_ID}")
print(f"\nProject details:")
project_details = client.projects.get_details(PROJECT_ID)
print(f"   Name: {project_details['entity']['name']}")
```

#### Link default space

## Parameter set 示範
**[[IBM watsonx.ai Python SDK] Parameter Sets 使用方式](https://ibm.github.io/watsonx-ai-python-sdk/v1.5.4/core_api.html#parameter-sets)**

### 第一種方法：列出所有 Parameter sets，找到想要的再去存取

```python
# 列出所有 Parameter sets
parameter_sets_list = client.parameter_sets.list()
parameter_sets_list
```

```python
# 拿回 Parameter set 資料
parameter_set_id = parameter_sets_list.iloc[0]['ID']
parameter_set_details = client.parameter_sets.get_details(parameter_set_id)
print("Parameter set 詳細資料：")
print(json.dumps(parameter_set_details, indent=2, ensure_ascii=False))
```

```python
# 顯示真正的 Parameters 和 Value sets
print("Parameter set 的 Parameters 和 Value sets：")
print(json.dumps(parameter_set_details['entity']['parameter_set'], indent=2, ensure_ascii=False))
```

### 第二種方法：直接用 parameter name 找到並存取想要的 parameter set

```python
# 指定 parameter name 並拿回 parameter set 資料

parameter_set_name = "pipeline_parameters"

parameter_set_id = client.parameter_sets.get_id_by_name(parameter_set_name)
parameter_set_details = client.parameter_sets.get_details(parameter_set_id)
print("Parameter set 詳細資料：")
print(json.dumps(parameter_set_details, indent=2, ensure_ascii=False))
```

```python
# 顯示真正的 Parameters 和 Value sets
print("Parameter set 主要的：")
print(json.dumps(parameter_set_details['entity']['parameter_set'], indent=2, ensure_ascii=False))
```
